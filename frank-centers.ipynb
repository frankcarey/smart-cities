```python
zips = data.load_zipcodes("./data/NYC_shapefiles/ZIP_CODE_040114.shp")
zips.head()
```

```python
parks = data.load_parks("./data/NYC_shapefiles/Parks Properties.geojson")
parks.head()
```

```python
n_parks={}
for i in zips.index:
    zipcode = zips.iloc[i]
    near = parks.intersects(zipcode.geometry)
    n = near.sum()
    #print(n)
    n_parks[zipcode['ZIPCODE']] = n
n_parks
```

```python
import pandas as pd
zipcount = pd.DataFrame(n_parks.items(), columns = ['ZIPCODE', 'parkcount'])
```
